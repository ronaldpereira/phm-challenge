```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
RANDOM_SEED = 1212
```

```python
train = pd.read_csv("../data/train_features.csv")
```

```python
train.drop(["Time_failure", "Time_diff", "Usage_on_failure"], axis=1, inplace=True)
```

```python
train.head()
```

```python
train = pd.get_dummies(train, columns=["Asset", "Reason", "Part"])
```

```python
train.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    train.drop("Failure", axis=1),
    train["Failure"],
    stratify=train["Failure"],
    test_size=0.333,
    random_state=RANDOM_SEED,
)
```

```python
model = RandomForestRegressor(
    n_estimators=300, n_jobs=-1, verbose=1, random_state=RANDOM_SEED
)
```

```python
model.fit(x_train, y_train)
```
